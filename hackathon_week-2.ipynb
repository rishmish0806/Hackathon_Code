```python
data=pd.read_csv("hacktrain.csv")
print(data.isnull().sum())
data2=pd.read_csv("hacktest.csv")
# data.head()
# data.info()

mean_data=data.groupby('class').mean()
mean_data

# data.isnull().sum()

class_means = data.groupby('class').mean()

# Fill missing values more efficiently
for cls in data['class'].unique():
    # Get indices of rows belonging to this class
    class_mask = data['class'] == cls
    # Fill missing values with class means
    data.loc[class_mask] = data.loc[class_mask].fillna(class_means.loc[cls])



from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data_process=data.iloc[:,3:]
data_process=pd.DataFrame(scaler.fit_transform(data_process))

drop_rows=[]
for i in range(data.shape[0]):
    for col in data_process.columns:
        if(data_process.iloc[i][col]>3 or data_process.iloc[i][col]<-3):
            drop_rows.append(i)
            break

data.drop(drop_rows,inplace=True)
data.head()
print("INFO")
data.info()
```
